```python
census_filtered = census_annual[census_annual["Census_Year"].isin([2006,2007, 2020,2025])]
```

```python
census_filtered
```

```python
census_pivot = census_filtered.pivot(
    index="Region",
    columns="Census_Year",
    values="Persons"
).reset_index()

census_pivot.columns.name = None
census_pivot.rename(
    columns={2006: "persons_2006",2007: "persons_2007", 2020: "persons_2020",2025:"census_all_2025"},
    inplace=True
)
```

```python
census_pivot
```

```python
census_pivot
```

```python
census_pivot["census_age_5_17"] = (
    census_pivot["persons_2020"] - census_pivot["persons_2007"]
)

census_pivot["census_adults"] = census_pivot["persons_2007"]
census_pivot["census_adults_2024"] = census_pivot["persons_2006"]

```

```python
census_pivot["census_adults"]
```

```python
uidai_clean = UIDAI_Demographic[[
    "state_std",
    "total_age_5_17",
    "total_age_17",
    "E_total_age_5_17",
    "E_total_age_18_greater"
]].rename(columns={
    "state_std": "Region",
    "total_age_5_17": "demo_under_18",
    "total_age_17": "demo_over_18",
    "E_total_age_5_17":"enrol_under_18",
    "E_total_age_18_greater":"enrol_over_18"
})
```

```python
final_df = census_pivot.merge(
    uidai_clean,
    on="Region",
    how="left"
)
```

```python
final_df = final_df[[
    "Region",
    "census_age_5_17",
    "census_adults",
    "census_all_2025",
    "census_adults_2024",
    "demo_under_18",
    "demo_over_18",
    "enrol_under_18",
    "enrol_over_18"
    
]]

print(final_df.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np

adult_df = final_df.dropna(subset=["census_adults", "demo_over_18", "enrol_over_18"])

regions = adult_df["Region"]
x = np.arange(len(regions))

plt.figure()
plt.bar(x - 0.25, adult_df["census_adults"], width=0.25, label="Census Adults")
plt.bar(x, adult_df["demo_over_18"], width=0.25, label="Demographic Adults")
plt.bar(x + 0.25, adult_df["enrol_over_18"], width=0.25, label="Enrollment Adults")

plt.yscale("log")
plt.xticks(x, regions, rotation=90)
plt.title("Adults: Census vs Demographic vs Enrollment (log scale)")
plt.legend()
plt.tight_layout()
plt.show()
```

Initial log-scale visualization revealed a consistent multi-order magnitude gap between population presence and actual enrollment across nearly all jurisdictions. While the census and demographic data maintain a near-perfect linear relationship, the adult enrollment data exhibits a critical decoupling. In the 'Bottom 3' states, the correlation between actual adult population and enrollment collapses to -0.44, confirming that in these regions, enrollment is no longer an indicator of population size, but rather a reflection of localized operational bottlenecks.

```python
child_df = final_df.dropna(subset=["census_age_5_17", "demo_under_18", "enrol_under_18"])

regions = child_df["Region"]
x = np.arange(len(regions))

plt.figure()
plt.bar(x - 0.25, child_df["census_age_5_17"], width=0.25, label="Census Children (5â€“17)")
plt.bar(x, child_df["demo_under_18"], width=0.25, label="Demographic Children (<18)")
plt.bar(x + 0.25, child_df["enrol_under_18"], width=0.25, label="Enrollment Children (<18)")

plt.yscale("log")
plt.xticks(x, regions, rotation=90)
plt.title("Children: Census vs Demographic vs Enrollment (log scale)")
plt.legend()
plt.tight_layout()
plt.show()
```

Parallel analysis of child demographics shows an almost identical log-scale profile to adults, suggesting that the initial 'reach' of the system was uniform across age groups. However, the correlation data indicates that child enrollment has now become the primary anchor for the entire system. In diverging states, adult enrollment correlates at 0.97 with child enrollment, while having a negative correlation with the actual adult census. This implies a 'Family-Unit Enrollment' bias: the system only successfully captures adults when they are forced into the ecosystem via their childrenâ€™s requirements (e.g., school admissions or mandatory biometrics).

```python
scaled_df["adult_divergence"] = (
    scaled_df["adult_demo_to_census"]
    - scaled_df["adult_enrol_to_census"]
)
```

To move beyond visual correlation, a Standardized Divergence Metric was calculated. By subtracting the scaled enrollment ratio from the scaled demographic ratio, we isolate regions where the population exists in the Census but is missing from the 2025 enrollment updates.Zero Score: Indicates a state where enrollment perfectly mirrors demographic reality (Saturation is likely true).High Positive Score: Indicates "Ghost Population"â€”citizens who are present in demographic estimates but are not being captured by the current operational infrastructure.Conclusion: This ranking provides the "Risk Map" required to challenge the $99.9\%$ universal coverage claim with localized, data-driven evidence.

```python
adult_risk_states = (
    scaled_df[["adult_divergence"]]
    .sort_values("adult_divergence", ascending=False)
)

adult_risk_states.head(10)
```

```python
final_df["coverage_gap_2025"]=final_df["census_adults"]-final_df["census_adults_2024"]-(1.5*final_df["enrol_over_18"])-((1/6)*final_df['enrol_under_18'])
```

```python
final_df
```

```python
final_df["coverage_gap_percent_2025"]=(final_df["coverage_gap_2025"]/(final_df["census_all_2025"]))*100
```

```python
final_df["coverage_percent_2025"]=100-final_df["coverage_gap_percent_2025"]
```

```python
final_df["percentage_coverage_gap_rate"] = final_df["coverage_gap_percent_2025"]-1
```

```python
gap_threshold_target = [2, 5,7.5,10,15,20,30]  # means 98, 95, 90
rows = []

for target in gap_threshold_target:
    temp = final_df.copy()
    
    temp["coverage_drop_percentage"] = 100 - target #where did it reach
    temp["year"] = ( 2026+ (target - temp["coverage_gap_percent_2025"])/ temp["percentage_coverage_gap_rate"]
    )
    
    rows.append(temp)

# Stack all rows together
final_df_expanded = pd.concat(rows, ignore_index=True)
```

```python
coverage_recovery_rates = [-0.01,-0.05 ,-0.1,-0.2,-0.3,-0.5, -0.8,-1,-1.5,-2]  # means 98, 95, 90
rows = []

for rate in coverage_recovery_rates:
    temp = final_df_expanded.copy()
    
    temp["coverage_recovery_rate"] = -rate
    temp["recovery_year"] = (2026 + (1 - temp["coverage_gap_percent_2025"])/ rate
    )
    
    rows.append(temp)

# Stack all rows together
final_df_expanded = pd.concat(rows, ignore_index=True)
```

```python
final_df=final_df_expanded
```

```python
final_df = final_df.drop(columns=["adult_demo_to_census","adult_enrol_to_census","child_demo_to_census","child_enrol_to_census"])
```
